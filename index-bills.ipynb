```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Your code here

# Pseudo-code outline

# Split the transcript into words
# Create a dictionary
# Iterate through the text of Macbeth
# Update word counts
# Create Bar Graph
# Include descriptive titles and labels
```

```python
macbeth.find('David Reed')
```

```python
macbeth_text = macbeth[16634+len('David Reed'):].lower().replace('\r',' ').replace('\n',' ').replace('.', ' ')\
               .replace(',',' ').replace('?',' ').replace(':',' ').replace(';',' ').replace('(','').replace(')','').replace('1','').replace('2','').replace('3','')
macbeth_text
```

```python
word_list1 = macbeth_text.split(" ")
word_list1[:24]
```

```python
word_list_all = None
while '' in word_list1:
    word_list1.remove('')
word_list_all = word_list1
word_list_all
```

```python
word_list_unique = []
for word in word_list1:
    if word not in word_list_unique:
        word_list_unique.append(word)
word_list_unique
```

```python
histogram_dict = dict.fromkeys(word_list_unique, None)
histogram_dict
```

```python
for k in histogram_dict.keys():
    histogram_dict[k] = word_list_all.count(k)
histogram_dict
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Set plot space as inline for inline plots and qt for external plots
%matplotlib inline
```

```python
sorted_histogram_tuplelist = sorted(histogram_dict.items(), key = lambda x: x[1], reverse = True)
sorted_histogram_tuplelist
```

```python
top_25 = sorted_histogram_tuplelist[:24]
top_25
```

```python
top_25_dict = dict(top_25)
top_25_dict
```

```python
plt.figure(figsize = (20,3))
plt.bar(top_25_dict.keys(), top_25_dict.values(), color='b')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Histogram of MacBeth Text')
plt.show()
```
